```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d as p3
from mpl_toolkits.mplot3d.art3d import juggle_axes
```

```python
nodes = np.load("data/loc_train_dance_short.npy")
velocities = np.load("data/vel_train_dance_short.npy")
edges = np.load("data/edges_train_dance_short.npy")

print(nodes.shape) # simulation_index, timesteps within the simulation (5000 timesteps sampled every 100 steps - 1 = 49 total) , x & y positions, node_index
print(velocities.shape) # simulation_index, timesteps within the simulation (5000 timesteps sampled every 100 steps - 1 = 49 total), x & y velocities, node_index
print(edges.shape) # simulation_index, node_index_start, node_index_end. This is fixed for the whole 49-step sequence. Note that the entries are either 0 or 1 -- 2 types of springs (i.e. spring or no spring)
```

```python
fig, ax = plt.subplots(1,3, figsize=(21,7))
fig.suptitle("Movement Phrases", fontsize=30)

for i in range(3):
    seq_number = i
    ax[i].plot(nodes[seq_number, :, 0, 0], nodes[seq_number, :, 1, 0], marker='o', markevery=6, label="Joint 0")
    ax[i].plot(nodes[seq_number, :, 0, 1], nodes[seq_number, :, 1, 1], marker='o', markevery=6, label="Joint 1")
    ax[i].plot(nodes[seq_number, :, 0, 2], nodes[seq_number, :, 1, 2], marker='o', markevery=6, label="Joint 2")
    ax[i].plot(nodes[seq_number, :, 0, 3], nodes[seq_number, :, 1, 3], marker='o', markevery=6, label="Joint 3")
    ax[i].plot(nodes[seq_number, :, 0, 4], nodes[seq_number, :, 1, 4], marker='o', markevery=6, label="Joint 4")
    ax[i].set_title("Sequence #{}".format(seq_number), fontsize=14)
    ax[i].tick_params(labelsize=12)
    ax[i].legend(fontsize=16)
```

```python
joint_names=['ARIEL','C7','CLAV','LANK','LBHD','LBSH','LBWT','LELB','LFHD','LFRM','LFSH','LFWT','LHEL','LIEL','LIHAND','LIWR','LKNE','LKNI','LMT1','LMT5','LOHAND','LOWR','LSHN','LTHI','LTOE','LUPA','MBWT','MFWT','RANK','RBHD','RBSH','RBWT','RELB','RFHD','RFRM','RFSH','RFWT','RHEL','RIEL','RIHAND','RIWR','RKNE','RKNI','RMT1','RMT5','ROHAND','ROWR','RSHN','RTHI','RTOE','RUPA','STRN','T10']
```

```python
fig = plt.figure(figsize=(25,7))
# ax = p3.Axes3D(fig)

joints = ['ARIEL','C7','CLAV','LANK','LBHD']
indices = [joint_names.index(joint) for joint in joints]

### Sequence 0
seq_number = 0
ax = fig.add_subplot(1,3,1, projection='3d')
for i in indices:
    ax.scatter(nodes[seq_number, :, 0, i],nodes[seq_number, :, 1, i],nodes[seq_number, :, 2, i], marker='o', 
#                label=joint_names[i]
               label="Joint "+str(i)
              )
ax.legend()
ax.set_title("Sequence {}".format(seq_number))


### Sequence 1
seq_number = 1
ax = fig.add_subplot(1,3,2, projection='3d')
for i in indices:
    ax.scatter(nodes[seq_number, :, 0, i],nodes[seq_number, :, 1, i],nodes[seq_number, :, 2, i], marker='o', 
#                label=joint_names[i]
               label="Joint "+str(i)
              )
ax.legend()
ax.set_title("Sequence {}".format(seq_number))


### Sequence 2
seq_number = 2
ax = fig.add_subplot(1,3,3, projection='3d')
for i in indices:
    ax.scatter(nodes[seq_number, :, 0, i],nodes[seq_number, :, 1, i],nodes[seq_number, :, 2, i], marker='o', 
#                label=joint_names[i]
               label="Joint "+str(i)
              )
ax.legend()
ax.set_title("Sequence {}".format(seq_number))

```
